**Se visualizan los datos duplicados mediante un grafico de barras**

```python
unicos0 = (coursera.nunique()).sort_values(ascending=False)
unicos1 = (reviews.nunique()).sort_values(ascending=False)
unicos2 = (edx.nunique()).sort_values(ascending=False)
unicos3 = (udemy.nunique()).sort_values(ascending=False)
```

```python
def graficarBar(x, y, titulo, titulo_x, titulo_y):

    '''
        Grafica un gráfico de barras
        x -> valores para el eje x
        y -> valores para el eje y
        titulo_x ->  titulo para eje x
        titulo_y ->  titulo para eje y

    '''

    plt.figure(figsize=(20,11))
    plt.title(titulo)
    plt.xlabel(titulo_x)
    plt.xticks(rotation=80)
    plt.ylabel(titulo_y)
    plots = sb.barplot(x=x, y=y, color='#4b00bb')
    #Agregamos el valor arriba de cada barrita
    for bar in plots.patches:     
        plots.annotate(format(bar.get_height(), '.2f'), 
                    (bar.get_x() + bar.get_width() / 2, 
                        bar.get_height()), ha='center', va='center', 
                    size=18, color='green', xytext=(0, 10), 
                    textcoords='offset points') 

    return plt.show()
```

```python
#COURSERA
graficarBar(unicos0.index, unicos0,'Valores Únicos en el Dataset', 'Columnas del Dataset', 'Cantidad de valores únicos')
```

```python
#COURSERA REVIEWS
graficarBar(unicos1.index, unicos1,'Valores Únicos en el Dataset', 'Columnas del Dataset', 'Cantidad de valores únicos')
```

```python
#EDX
graficarBar(unicos2.index, unicos2,'Valores Únicos en el Dataset', 'Columnas del Dataset', 'Cantidad de valores únicos')
```

```python
#UDEMY
graficarBar(unicos3.index, unicos3,'Valores Únicos en el Dataset', 'Columnas del Dataset', 'Cantidad de valores únicos')
```

**Se nos solicita un WordCloud de las palabras clave que más se repiten dentro del título**

```python
#Se importa las librerias a utlizar
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
#Se crea la primera imagen para edx_n
text0 = coursera_merge.title.values
wordcloud = WordCloud(
    width = 300,
    height = 200,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text0))
fig = plt.figure(
    figsize = (20, 10),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig("courseraword.png")
plt.show()
```
